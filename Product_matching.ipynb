```python
# Features selection 
data1_features = ['uniq_id','product_name','retail_price','discounted_price','description']
```

```python
data2_features = ['uniq_id','product_name','retail_price','discounted_price','description']
```

### **Dealing with null values**

```python
# replacing nul value with null strng
for feature in data1_features:
  data1[feature] = data1[feature].fillna('')

# replacing nul value with null strng
for feature in data2_features:
  data2[feature] = data2[feature].fillna('')
```

```python
a = data1[data1["product_name"]=="Alisha Solid Women's Cycling Shorts"]
b = a['uniq_id'][0]
e = a.head(1)
flipcart_product_name = e['product_name']
flipcart_product_price = e['retail_price']
flipcart_product_discountd_price = e['discounted_price']
c = data2[data2["uniq_id"]==b]
amazone_product_name = c['product_name']
amazone_product_price = c['retail_price']
amazone_product_discountd_price = c['discounted_price']
from prettytable import PrettyTable
train = PrettyTable(["flipcart product name", "flipcart product price", " flipcart product discounted price","amazone product name", "amazone product price", " amazone product discounted price"])
train.add_row([flipcart_product_name,flipcart_product_price,flipcart_product_discountd_price,amazone_product_name,amazone_product_price,amazone_product_discountd_price])

print(train)
```
